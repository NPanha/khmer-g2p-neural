```python
# Quick stats + phoneme frequency bar chart.
from khmer_g2p.lexicon import load_tsv
from khmer_g2p.neural.phoneme_tokenizer import tokenize as phoneme_tokenize

pairs = load_tsv(str(DATA_PATH))
words = [w for w, _ in pairs]
ipas  = [p for _, p in pairs]
word_lens   = [len(w) for w in words]
ipa_lens    = [len(phoneme_tokenize(p)) for p in ipas]
phone_freq  = Counter(t for p in ipas for t in phoneme_tokenize(p))

print(f"entries:       {len(pairs):,}")
print(f"unique words:  {len(set(words)):,}")
print(f"word chars:    mean={np.mean(word_lens):.1f}  p95={np.percentile(word_lens,95):.0f}")
print(f"IPA phonemes:  mean={np.mean(ipa_lens):.1f}  p95={np.percentile(ipa_lens,95):.0f}")
print(f"unique phonemes: {len(phone_freq)}")

fig, axes = plt.subplots(1, 3, figsize=(15, 3.5))
axes[0].hist(word_lens, bins=30, color='steelblue'); axes[0].set_title('Word length (chars)')
axes[1].hist(ipa_lens, bins=30, color='seagreen'); axes[1].set_title('IPA length (phonemes)')
top = phone_freq.most_common(30)
axes[2].barh([t for t,_ in top[::-1]], [c for _,c in top[::-1]], color='coral')
axes[2].set_title('Top 30 phonemes')
plt.tight_layout(); plt.show()
```
